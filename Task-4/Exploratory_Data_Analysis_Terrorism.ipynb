```python
data['Wounded'].value_counts()
```

```python
state = data.groupby('Country')[['Wounded']].sum().sort_values(by=['Wounded'], ascending=False)
state
```

```python
#state = data['Wounded'].value_counts()[:20]
state[:20].plot.bar(color = ["indigo"], figsize=(20,10))

plt.xlabel("Country", fontsize=15)
plt.ylabel("Wounded", fontsize=15)
plt.xticks(rotation =65)
plt.title("Wounded person in different countries", fontsize=20)
```

```python
year = data.groupby('Year')[['Wounded']].sum().sort_values(by=['Wounded'], ascending=False)
year
```

```python
year.plot.bar(color = ["indigo"], figsize=(20,10))

plt.xlabel("Year", fontsize=15)
plt.ylabel("Wounded", fontsize=15)
plt.xticks(rotation =65)
plt.title("Wounded person in different Years", fontsize=20)
```

```python
year = data.groupby('Year')[['Wounded']].sum()
year.plot(color = ["red"], figsize=(10,5), kind='line', marker='o')
plt.xlabel("Year", fontsize=15)
plt.ylabel("Wounded", fontsize=15)
plt.xticks(rotation =65, fontsize=15)
plt.yticks(fontsize=15)
plt.title("Wounded person in different Years", fontsize=20)
```

```python
year = data.groupby('Year')[['Killed']].sum()
year.plot(color = ["green"], figsize=(10,5), kind='line', marker='*')
plt.xlabel("Year", fontsize=15)
plt.ylabel("Killed", fontsize=15)
plt.xticks(rotation =65, fontsize=15)
plt.yticks(fontsize=15)
plt.title("Total person killed in different years", fontsize=20)
```
