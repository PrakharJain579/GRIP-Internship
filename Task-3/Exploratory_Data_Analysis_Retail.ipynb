```python
state = data.groupby('State')[['Profit']].sum().plot.bar(figsize=(30,10))
plt.xticks(rotation =65)
plt.xlabel("State", fontsize=15)
plt.ylabel("Profit", fontsize=15)
plt.title("State wise Profit", fontsize=20)
plt.xticks(fontsize = 15, rotation =65)
plt.yticks(fontsize = 15)
```

```python
state = data.groupby('Category')[['Sales']].sum().plot.bar(figsize=(10,10))
plt.xticks(rotation =65)
plt.ylabel("Sales", fontsize = 15)
plt.xlabel("Category", fontsize = 15)
plt.title("Sales in different categories", fontsize = 20)
```

**Here we can observe that Technology has more sales than furniture and at last office supplies.**

```python
state = data.groupby('Sub-Category', sort = True)[['Sales']].sum()
state = state.sort_values(['Sales'], ascending = False)
state.plot.bar(color = ["magenta"], figsize=(20,10))

plt.xlabel("Sub-Categories", fontsize=15)
plt.ylabel("Sales", fontsize=15)
plt.xticks(rotation =65)
plt.title("Sales in different sub-categories", fontsize=20)
```

**From this graph we can conclude that Phones has the highest sales then chairs and so on.**

```python
state = data.groupby('Region')[['Sales']].sum()
state = state.sort_values(['Sales'], ascending = False)
state.plot.bar(color = ["indigo"], figsize=(10,10))

plt.xlabel("Region", fontsize=15)
plt.ylabel("Sales", fontsize=15)
plt.xticks(rotation =65)
plt.title("Sales in different Region", fontsize=20)
```

**This shows that west region has heighest sale and south region has lowest.**

```python
state = data.groupby('Sub-Category', sort = True)[['Sales']].sum()
state.plot(kind='line', color = ["magenta"], figsize=(20,10))
plt.title("Sales",fontsize= 14)
plt.xlabel("Sub-Category", size = 14)
plt.ylabel("Sales ", size = 14)
plt.xticks(size = 14)
plt.yticks(size = 14)
```

***From above data analysis some conclusions we can drive are:***


*   Standard Class ship mode is prefered more and  same day ship mode is preffered least.
*   There is high demand for consumer segment and least demand for Home office segment.
*   States like California and New-York has highest profit whereas Texas has least profit soe selling our products in these two states will benefit us more than any other states.
*   Sub categories like phones, tables, chairs, binders has highest sales whereas fateners, labels, envelpoes etc has has the least sales so we need to focus and promote these categories more to increase there sales.
*   West region has highest sales where as South region has lowest so by selling our products in western region will be benificial and we need to take some measures to increase sales in south region.
